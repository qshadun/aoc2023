```python
from collections import defaultdict
import random

# for test
test_input = '../../inputs/test25.txt'

# for real input
my_input = '../../inputs/input25.txt'

class Node:
    def __init__(self):
        self.count = 1

def get_index(name, name2idx):
    if name not in name2idx:
        name2idx[name] = len(name2idx)
    return name2idx[name]

def part1(input_file):
    name2idx = {}
    src_edges = []
    for line in open(input_file):
        parts = line.split(': ')
        from_node = parts[0].strip()
        from_idx = get_index(from_node, name2idx)
        to_nodes = [x.strip() for x in parts[1].split(' ')]
        for to_node in to_nodes:
            to_idx = get_index(to_node, name2idx)
            src_edges.append((min(from_idx, to_idx), max(from_idx, to_idx)))
    n = len(name2idx)
    while True:
        nodes = [Node() for _ in range(n)]
        edges = src_edges.copy()
        node_count = n
        while node_count > 2:
            idx = random.randint(0, len(edges) - 1)
            a, b = edges[idx]
            nodes[a].count += nodes[b].count
            node_count -= 1
            i = 0
            while i < len(edges):
                c, d = edges[i]
                if a == c and b == d:
                    if i == len(edges) - 1:
                        edges.pop()
                    else:
                        edges[i] = edges.pop()
                elif c == b:
                    edges[i] = (min(a, d), max(a, d))
                    i += 1
                elif d == b:
                    edges[i] = (min(a, c), max(a, c))
                else:
                    i += 1
        if len(edges) == 3:
            a, b = edges[0]
            return nodes[a].count * nodes[b].count
```
