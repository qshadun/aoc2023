```python
import graphviz
import os
from collections import defaultdict
```

```python
graph = defaultdict(list)
# for test
# input_file = '../../inputs/input25.txt'
# graph_name = 'machines'
# from dot plot, we know this two nodes belong to two partitions
# start_node = 'bvb'
# end_node = 'cmg'

# for real input
input_file = '../../inputs/input25.txt'
graph_name = 'machines'
start_node = 'bvc'
end_node = 'rsm'

def part1():
    for line in open(input_file):
        parts = line.split(': ')
        from_node = parts[0].strip()
        to_nodes = [x.strip() for x in parts[1].split(' ')]
        for to_node in to_nodes:
            graph[from_node].append(to_node)
            graph[to_node].append(from_node)
part1()
n = len(graph)
graph
```

```python
dot = graphviz.Graph(graph_name)
for k in graph:
    dot.node(k)

for line in open(input_file):
    parts = line.split(': ')
    from_node = parts[0].strip()
    to_nodes = [x.strip() for x in parts[1].split(' ')]
    for to_node in to_nodes:
        dot.edge(from_node, to_node)
```

```python
name_to_index = {}
cur = 0
for k in graph:
    name_to_index[k] = cur
    cur += 1
print(name_to_index)
```

```python
matrix = [[0] * len(graph) for _ in range(len(graph))]
for left, rs in graph.items():
    for r in rs:
        li = name_to_index[left]
        ri = name_to_index[r]
        matrix[li][ri] = 1
        matrix[ri][li] = 1
 
```

```python
dot1 = graphviz.Graph(graph_name + '_num')
for k in range(n):
    dot1.node(str(k))

for i in range(n-1):
    for j in range(i + 1, n):
        if matrix[i][j]:
            dot1.edge(str(i), str(j))
dot1.render(directory='.', view=True)
```

```python
start = name_to_index[start_node]
end = name_to_index[end_node]
print(matrix[start][end])
matrix[start][end] = 0
matrix[end][start] = 0

all_edges = []
for i in range(n - 1):
    for j in range(i+1, n):
        if matrix[i][j]:
            all_edges.append((i, j))
print(all_edges)
```

```python
print(start, end, n, len(all_edges), matrix[start][end], matrix[end][start])
```

```python
from collections import deque, Counter

for i in range(len(all_edges) - 1):
    l1, r1 = all_edges[i]
    if l1 == start or l1 == end or l1 == start or r1 == end:
        continue
    if not matrix[l1][r1] or not matrix[r1][l1]:
            print("error l1 r1", l1, r1)
    matrix[l1][r1] = 0
    matrix[r1][l1] = 0
    found_end = False
    for j in range(i + 1, len(all_edges)):
        
        l2, r2 = all_edges[j]
        if l2 == start or l2 == end or l2 == start or r2 == end:
            continue
        if not matrix[l2][r2] or not matrix[r2][l2]:
            print("error l2, r2", l2, r2)
        matrix[l2][r2] = 0
        matrix[r2][l2] = 0
        visited = [0] * n
        visited[start] = 1
        q = deque()
        q.append(start)
        found_end = False
        while q:
            cur = q.popleft()
            if cur == end:
                found_end = True
                break
            for k in range(n):
                if matrix[cur][k] and not visited[k]:
                    visited[k] = 1
                    q.append(k)
        # succeed
        if not found_end:
            n1= Counter(visited)[1]
            print("succeed: ", n1 * (n - n1))
            break
        # print(i, j, all_edges[i], all_edges[j], visited)
        matrix[l2][r2] = 1
        matrix[r2][l2] = 1
    if not found_end:
        break
    matrix[l1][r1] = 1
    matrix[r1][l1] = 1

                    


        
```
