```python
import graphviz
import os
from collections import defaultdict
```

```python
graph = defaultdict(list)
# input_file = 'inputs/test25.txt'
# graph_name = 'machines_test'
input_file = 'inputs/input25.txt'
graph_name = 'machines'
def part1():
    for line in open(input_file):
        parts = line.split(': ')
        from_node = parts[0].strip()
        to_nodes = [x.strip() for x in parts[1].split(' ')]
        for to_node in to_nodes:
            graph[from_node].append(to_node)
            graph[to_node].append(from_node)
part1()
graph
```

```python
dot = graphviz.Graph(graph_name)
for k in graph:
    dot.node(k)

for line in open(input_file):
    parts = line.split(': ')
    from_node = parts[0].strip()
    to_nodes = [x.strip() for x in parts[1].split(' ')]
    for to_node in to_nodes:
        dot.edge(from_node, to_node)
```

```python
dot.render(directory='.', view=True)
```
